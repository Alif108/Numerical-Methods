# Exponential Function

The exponential function is defined as:

$$ f(x) = e^x $$

where $e$ is Euler's number, approximately $2.71828$.

```python
# plot the exponential function for x in the interval [0, 10]
x = np.arange(0, 10)
y = np.exp(x)

plt.figure(figsize=(10, 5))
plt.title("Exponential function")
plt.xlabel("n")
plt.ylabel("Exp(n)")
plt.plot(x, y)
plt.show()
```

# Question

### Which grows faster? $x^2$ or $2^x$?

```python
# plot x^2 and 2^x on the same graph

x = np.linspace(0, 5, 100)
y1 = x**2
y2 = 2**x

plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='$x^2$', color='orange', marker='o', linewidth=0.5)
plt.plot(x, y2, label='$2^x$', color='blue', marker='x', linewidth=0.5)
plt.legend()
plt.show()
```
