### Plot the following functions in the range [-1, 1] with 1000 points:
### 1. $y = x$
### 2. $y = x^2$
### 3. $y = x^3$
### 4. $y = e^x$
### 5. $y = tan(x)$
in the same plot. Use different colors for each function and add a legend to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 1000)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.exp(x)
y5 = np.tan(x)

plt.figure(figsize=(10, 10))
plt.plot(x, y1, label='x')
plt.plot(x, y2, label='x^2')
plt.plot(x, y3, label='x^3')
plt.plot(x, y4, label='e^x')
plt.plot(x, y5, label='tan(x)')
plt.legend()
plt.show()
```

### Plot the following function in the range [0, 2] with 1000 points:
$$ x = 16sin^3(\theta) $$
$$ y = 13cos(\theta) - 5cos(2\theta) - 2cos(3\theta) - cos(4\theta) $$

```python
theta = np.linspace(0, 2*np.pi, 1000)

x = 16 * np.sin(theta)**3
y = 13 * np.cos(theta) - 5 * np.cos(2*theta) - 2 * np.cos(3*theta) - np.cos(4*theta)

plt.figure(figsize=(10, 10))
plt.plot(x, y, color='red', linewidth=3)
plt.fill(x, y, color='red', alpha=0.2)
plt.title('Heart Curve', fontsize=20)
plt.show()
```

### Plot the following function in the range [0, 2] with 1000 points:
$$ x = sin(2\theta) $$
$$ y = sin(3\theta) $$

```python
theta = np.linspace(0, 2*np.pi, 1000)
x = np.sin(2*theta)
y = np.cos(3*theta)

plt.figure(figsize=(10, 10))
plt.plot(x, y)
plt.show()
```

### Plot the following function in the range [0, 300] with 1000 points:
$$ x = -35cos(\theta) + 65cos(-0.35\theta) $$
$$ y = -35sin(\theta) + 65sin(-0.35\theta) $$

```python
theta = np.linspace(0, 300, 1000)
x = -35*np.cos(theta) + 65*np.cos(-0.35*theta)
y = -35*np.sin(theta) - 65*np.sin(-0.35*theta)

plt.figure(figsize=(10, 10))
plt.plot(x, y)
plt.show()
```

# The Butterfly Curve
### Plot the following function in the range [0, 12] with 1000 points:
$$ x = sin\theta (e^{cos\theta} - 2cos(4\theta) - sin^5(\theta/12)) $$
$$ y = cos\theta (e^{cos\theta} - 2cos(4\theta) - sin^5(\theta/12)) $$

```python
t = np.linspace(0, 12*np.pi, 1000)
x = np.sin(t) * ( np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12)**5 )
y = np.cos(t) * ( np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12)**5 )

plt.figure(figsize=(10,10))
plt.plot(x, y)
plt.title("Butterfly Curve", fontsize=20)
plt.show()
```

# Plot from Table
Consider the following table : 

```python
stats = [[2, 78],
         [5, 93],
         [1.5, 70],
         [6.5, 72],
         [5.5, 84]]

# sort the list based on the first element of each sublist
stats.sort(key=lambda x: x[0])

stats = np.array(stats)
plt.xlabel('Hours')
plt.ylabel('Score')
plt.plot(stats[:, 0], stats[:, 1])
plt.show()
```
