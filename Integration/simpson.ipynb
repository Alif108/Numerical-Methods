```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simpsons_one_third(f, a, b, n):
    h = (b - a)/n
    x = np.linspace(a, b, n+1)
    y = f(x)
    return h/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])
```

```python
def func(x):
    return np.exp(-x**2)

# plot the function
x = np.linspace(0, 3, 100)
y = func(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = exp(-x^2)')
plt.show()
```

```python
integral = simpsons_one_third(func, 1, 2, 100)
print('The integral of exp(-x^2) from 1 to 2 is', integral)

# mark the area under the curve
x = np.linspace(0, 3, 100)
y = func(x)
plt.plot(x, y)

x1 = np.linspace(1, 2, 100)
y1 = func(x1)
plt.fill_between(x1, y1)

plt.xlabel('x')
plt.ylabel('y')
plt.title('$\int_1^2 e^{-x^2} dx$')
plt.show()
```
