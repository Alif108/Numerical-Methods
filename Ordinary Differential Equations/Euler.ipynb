```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f_prime(x, y):
    return 2*x
```

```python
def euler(x0, y0, h, n, f_prime):
    x = x0
    y = y0

    xs = [x0]
    ys = [y0]

    for i in range(n):
        y += h * f_prime(x, y)
        x += h
        print("x = ", x, "y = ", y, "f'(x) = ", f_prime(x, y))

        xs.append(x)
        ys.append(y)

    return xs, ys, y
```

```python
xs, ys, eu = euler(0, 1, 1, 5, f_prime)
print(eu)
```

```python
# plot y=x^2 and xs and ys in the same plot
x = np.linspace(0, 5, 100)
y = x**2
plt.plot(x, y, label="y = x^2")
plt.plot(xs, ys, label="euler")
plt.legend()
plt.show()
```
