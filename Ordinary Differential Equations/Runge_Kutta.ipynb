```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dydx(x,y):
    return -2.2067E-12 * (y**4 - 81E8)
    # return y
```

```python
def runge_kutta_heun(x0,y0,xn,n):
    h = (xn-x0)/n
    x = np.linspace(x0,xn,n+1)
    y = np.zeros([n+1])
    y[0] = y0
    for i in range(n):
        k1 = dydx(x[i],y[i])
        k2 = dydx(x[i]+h,y[i]+h*k1)
        y[i+1] = y[i] + (h/2)*(k1+k2)
    return x,y

def runge_kutta_ralston(x0,y0,xn,n):
    h = (xn-x0)/n
    x = np.linspace(x0,xn,n+1)
    y = np.zeros([n+1])
    y[0] = y0
    for i in range(n):
        k1 = dydx(x[i],y[i])
        k2 = dydx(x[i]+(3/4)*h,y[i]+(3/4)*h*k1)
        y[i+1] = y[i] + (h/3)*(k1+2*k2)
    return x,y

def runge_kutta_midpoint(x0,y0,xn,n):
    h = (xn-x0)/n
    x = np.linspace(x0,xn,n+1)
    y = np.zeros([n+1])
    y[0] = y0
    for i in range(n):
        k1 = dydx(x[i],y[i])
        k2 = dydx(x[i]+(1/2)*h,y[i]+(1/2)*h*k1)
        y[i+1] = y[i] + h*k2
    return x,y
```

```python
# call the function
x0 = 0
xn = 480
y0 = 1200
n = 2
x,y = runge_kutta_heun(x0,y0,xn,n)
plt.plot(x,y,'o-',label='Heun')
x,y = runge_kutta_ralston(x0,y0,xn,n)
plt.plot(x,y,'o-',label='Ralston')
x,y = runge_kutta_midpoint(x0,y0,xn,n)
plt.plot(x,y,'o-',label='Midpoint')

plt.legend()
plt.show()
```
